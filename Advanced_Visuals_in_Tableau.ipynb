**Question 4 : Create a Butterfly Chart comparing Male vs Female patient count by Department.**

**ANS:**

image.png

A butterfly chart was created by plotting Female patient counts on the positive axis and Male patient counts as negative values. This allows a mirrored comparison across departments. The chart clearly shows that female patient counts are generally higher than male patients across most departments, with Pediatrics and Neurology having the highest counts. This visualization helps in identifying gender-based distribution patterns in hospital departments.
